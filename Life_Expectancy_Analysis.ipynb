```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler



# ðŸ” Clean column names (remove spaces)
df.columns = df.columns.str.strip()

# âœ… Select 5 features and target
features = ['Status', 'Adult Mortality', 'Alcohol', 'BMI', 'Schooling']
df_model = df[features + ['Life expectancy']].dropna()

# Features and Target
X = df_model[features]
y = df_model['Life expectancy']

# Scale features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=features)

```

```python
import joblib

# Save scaler and column names
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(X_scaled_df.columns.tolist(), 'feature_columns.pkl')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42)

# Train
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"âœ… RÂ² Score: {r2:.4f}")
print(f"âœ… RMSE: {rmse:.4f}")

# Save model
joblib.dump(model, 'life_expectancy_model.pkl')
```

```python
# Load saved model, scaler, and column order
model = joblib.load('life_expectancy_model.pkl')
scaler = joblib.load('scaler.pkl')
feature_columns = joblib.load('feature_columns.pkl')

# Dummy input example (in same order as feature_columns)
dummy = pd.DataFrame([[1, 160, 4.5, 22.0, 12.0]], columns=feature_columns)

# Scale and predict
scaled_dummy = scaler.transform(dummy)
prediction = model.predict(scaled_dummy)

print("âœ… Predicted Life Expectancy:", prediction[0])

```

### Linear Regressor Model
After training the Linear Regression model to predict Life Expectancy, we evaluate its performance using two key regression metrics:
RÂ² Score (Coefficient of Determination): Indicates how well the model explains the variance in the target variable. A value closer to 1 means better performance.
Root Mean Squared Error (RMSE): Measures the average prediction error. Since we used MinMaxScaler, RMSE is normalized between 0 and 1.
